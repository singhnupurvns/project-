```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df[["age","affordibility"]],df.bought_insurance,test_size =0.2,random_state =25)
```

```python
x_train_scaled = x_train.copy()
x_train_scaled["age"] = x_train_scaled["age"]/100
x_test_scaled = x_test.copy()
x_test_scaled["age"] = x_test_scaled["age"]/100
```

```python
x_train_scaled.head()
```

```python
model = keras.Sequential([
       keras.layers.Dense(1,input_shape = (2,),activation = "sigmoid",kernel_initializer = "ones",bias_initializer= "zeros")])

model.compile(optimizer = "adam",loss = "binary_crossentropy",metrics = ["accuracy"])
model.fit(x_train_scaled,y_train,epochs = 2000)
```

```python
model.evaluate(x_test_scaled,y_test)
```

```python
def log_loss(y_true,y_predicted):
    epsilon = 1e-15
    y_predicted_new =[max(i,epsilon)for i in y_predicted]
    y_predicted_new1=[min(i,1-epsilon)for i in y_predicted_new]
    y_predicted_new2 = np.array(y_predicted_new1)
    return -np.mean(y_true*np.log(y_predicted_new2)+(1-y_true)*np.log(1-y_predicted_new2))
```
